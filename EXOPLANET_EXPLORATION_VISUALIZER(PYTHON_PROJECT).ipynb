```python
# CONDITION, FUNCTION, DICTIONARY , LOOP
# TO CHECK WHETHER AN EXOPLANET IS HABITABLE OR NOT
Exoplanets = [
    {"name": "Gas Giants", "distance_from_star": 10},
    {"name": "Neptunian planets", "distance_from_star": 30},
    {"name": "super earths", "distance_from_star": 2},
    {"name": "Terrestrial planets", "distance_from_star": 2.5},
]

def is_habitable(dis_from_star):
  habitable_zone_min = 0.95
  habitable_zone_max = 3.0

  if habitable_zone_min <= dis_from_star < habitable_zone_max:
    return True
  else:
    return False

# Check each exoplanet for habitability
for planet in Exoplanets:
    name = planet["name"]
    distance = planet["distance_from_star"]

    if is_habitable(distance):
        print(f"{name} is within the habitable zone!")
    else:
        print(f"{name} is NOT within the habitable zone.")
```

```python
# CLASSES AND SUBCLASSES
# CLASS
class exoplanets:
  def __init__(self, name , distance_from_star,temp, habitability , star_type):
    self.name = name
exoplanets1 = exoplanets("gas giants" , "10", -180, "no", "G-Type")
exoplanets2 = exoplanets("neptunian planets" , "30" , -214, "no" ,"G-Type")
exoplanets3 = exoplanets("super earths" , "2" , 50 , "yes", "G-Type")
exoplanets4 = exoplanets("terrestrial planets" , "2.5" , 14 ,"yes" , "G-Type")


# SUBCLASSES
# Gas Giants
class gas_giants(exoplanets):
  def __init__(self, name, distance_from_star, temp, habitability, star_type):
    super().__init__(name, distance_from_star, temp, habitability, star_type)
    self.temp = temp
    self.habitability = habitability
    self.star_type = star_type

gas_giant_example = gas_giants("Gas Giant ", 10, -180, "no", "G-Type")
print(gas_giant_example.name)
print(gas_giant_example.temp)
print(gas_giant_example.habitability)
print(gas_giant_example.star_type)

# Neptunian planets
class neptunian_planets(exoplanets):
  def __init__(self, name , distance_from_star , temp , habitability , star_type):
    super().__init__(name , distance_from_star , temp , habitability , star_type)
    self.temp = temp
    self.habitability = habitability
    self.star_type = star_type

neptunian_planet_example = neptunian_planets("neptunian planets" , "30" , -214, "no" ,"G-Type")
print(neptunian_planet_example.name)
print(neptunian_planet_example.temp)
print(neptunian_planet_example.habitability)
print(neptunian_planet_example.star_type)

# Super Earths
class super_earths(exoplanets):
  def __init__(self, name , distance_from_star, temp , habitability , star_type):
    super().__init__(name , distance_from_star , temp , habitability , star_type)
    self.temp = temp
    self.habitability = habitability
    self.star_type = star_type
super_earths_example = super_earths("super earths" , "2" , 50 , "yes", "G-Type")
print(super_earths_example.name)
print(super_earths_example.temp)
print(super_earths_example.habitability)
print(super_earths_example.star_type)

# Terrestrial planets
class terrestrial_planets(exoplanets):
  def __init__(self, name , distance_from_star , temp , habitability , star_type):
    super().__init__(name , distance_from_star , temp , habitability , star_type)
    self.temp = temp
    self.habitability = habitability
    self.star_type = star_type
terrestrial_planets_example = terrestrial_planets("terrestrial planets" , "2.5" , 14 ,"yes" , "G-Type")
print(terrestrial_planets_example.name)
print(terrestrial_planets_example.temp)
print(terrestrial_planets_example.habitability)
print(terrestrial_planets_example.star_type)

```

```python
# LIBRARIES : VISUALIZATION OF EXOPLANETS
import matplotlib.pyplot as plt

# Exoplanet data (Distance from star, Size, Habitability)
planets_data = [
    {"name": "Super Earth", "distance": 2, "size": 4},
    {"name": "Terrestrial Planet", "distance": 2.5, "size": 2},
    {"name": "Neptunian Planet", "distance": 30, "size": 3},
    {"name": "Gas Giant", "distance": 10, "size": 6}
]

# Create a simple plot
plt.figure(figsize=(8, 9))

# Plot each planet as a circle
for planet in planets_data:
    plt.scatter(planet["distance"], planet["size"], s=planet["size"]*100, label=planet["name"])

# Add title and labels
plt.title("Simple Exoplanet Visualization")
plt.xlabel("Distance from Star (AU)")
plt.ylabel("Planet Size (Relative to Earth)")

# Show the plot with legend
plt.legend()
plt.show()
```
